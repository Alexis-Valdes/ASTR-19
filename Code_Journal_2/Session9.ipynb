1. Create a Jupyter notebook, import matplotlib.
2. Use numpy to pull 1000 random numbers distributed uniformly between [0,1].
3. Histogram the random numbers into 100 bins, and plot the histogram.
4. Label your axes and save the figure as a PDF.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.uniform(0, 1, 1000)
width = 0.5

plt.hist(x, bins=100, alpha=0.5, edgecolor="black")
plt.ylabel("N per bin")
plt.xlabel("X")
plt.title("Uniform distrubution")
plt.savefig("Session9.pdf",bbox_inches = "tight" , facecolor = "white")
```
