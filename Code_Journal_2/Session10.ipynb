```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
size = 10. # shape
samples = 1000
s = np.random.power(size, samples)

plt.hist(s, bins=100, alpha=0.5, edgecolor="black")

plt.xlabel('X value')
plt.ylabel('# of points')
plt.title("Power Distrubution")
plt.savefig("session10.pdf", dpi=300, bbox_inches="tight", facecolor="white")

plt.show()
```
